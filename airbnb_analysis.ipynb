# Map-Based Analysis

this part is where I'll learn something new, I wanted to see and learn where I can turn this #latitude and #longitude into insights, and then while browsing I came to know about (tada) "Foliummm!!!!" (*Ahem. Folium) used for map visualiaztion, so I installed it with the command : "pip install folium" already

```python
import folium
from folium.plugins import MarkerCluster
```

so, I need to "Create a Base Map" first

```python
df[['latitude', 'longitude']].mean()
```

```python
albany_map = folium.Map(
    location=[df['latitude'].mean(), df['longitude'].mean()],
    zoom_start=12
)

albany_map
```

An interactive map

```python
marker_cluster = MarkerCluster().add_to(albany_map)

for _, row in df.iterrows():
    folium.Marker(
        location=[row['latitude'], row['longitude']],
        popup=f"Price: ${row['price']}"
    ).add_to(marker_cluster)

albany_map
```

now we can see 
>Density of listings

>Spatial clustering

>Popular hosting zones

Color-Code by Price (now we do the thang)

```python
def price_color(price):
    if price < 75:
        return 'green'
    elif price < 150:
        return 'orange'
    else:
        return 'red'
```

```python
price_map = folium.Map(
    location=[df['latitude'].mean(), df['longitude'].mean()],
    zoom_start=12
)

for _, row in df.iterrows():
    folium.CircleMarker(
        location=[row['latitude'], row['longitude']],
        radius=5,
        color=price_color(row['price']),
        fill=True,
        fill_opacity=0.7,
        popup=f"""
        <b>Price:</b> ${row['price']}<br>
        <b>Room Type:</b> {row['room_type']}<br>
        <b>Reviews:</b> {row['number_of_reviews']}
        """
    ).add_to(price_map)

price_map
```

```python
price_map.save("albany_airbnb_price_map.html")
```

ðŸŸ¢ Green: Budget stays (< $75)
ðŸŸ  Orange: Mid-range ($75â€“150)
ðŸ”´ Red: Premium stays (> $150)

```python
# Dark mode base map
price_map = folium.Map(
    location=[df['latitude'].mean(), df['longitude'].mean()],
    zoom_start=12,
    tiles="CartoDB dark_matter"
)

for _, row in df.iterrows():
    folium.CircleMarker(
        location=[row['latitude'], row['longitude']],
        radius=5,
        color=price_color(row['price']),
        fill=True,
        fill_color=price_color(row['price']),
        fill_opacity=0.75,
        popup=folium.Popup(
            f"""
            <div style="font-size:13px;">
                <b>Price:</b> ${row['price']}<br>
                <b>Room Type:</b> {row['room_type']}<br>
                <b>Reviews:</b> {row['number_of_reviews']}
            </div>
            """,
            max_width=250
        )
    ).add_to(price_map)

price_map
```

observation: 

High-price listings cluster in specific areas

Entire homes dominate premium zones

Also, Dense clusters means (probably) popular neighborhoods hmmmm
